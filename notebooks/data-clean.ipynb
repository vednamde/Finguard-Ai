```python
df.columns = (
    df.columns
    .str.strip()
    .str.replace("[()]","", regex = True)
)

df["Market Capin B USD"].fillna(df["Market Capin B USD"].median(), inplace=True)


df.to_csv("../data/Cleaned_Financial_Statements.csv", index=False)

print("Cleaned data saved successfully as 'Cleaned_financial_Statements.csv'")
```

```python
# =======================================================
# STEP 2: FEATURE ENGINEERING FOR FINANCIAL HEALTH MODEL
# =======================================================


df = pd.read_csv("../data/Cleaned_Financial_Statements.csv")


#create a Target Variable
df['Financial_Status'] = df['Net Income'].apply(lambda x:1 if x < 0 else 0)


# ------ Profitability Ratios ------
df['Profit_Margin_Ratio'] = df['Net Income'] / df['Revenue']
df['EBITDA_Margin'] = df['EBITDA'] / df['Revenue']
df['ROA_Ratio'] = df['ROA'] / 100
df['ROE_Ratio'] = df['ROE'] / 100


# ------ Liquidity Ratios ------
df['CashFlow_to_Revenue'] = df['Cash Flow from Operating'] / df['Revenue']
df['Liquidity_Index'] = df['Current Ratio'] * df['CashFlow_to_Revenue']



# ----- Efficiency Ratios -----
df['Profit_to_CashFlow'] = df['Net Income'] / df['Cash Flow from Operating']
df['Revenue_per_Employee'] = df['Revenue'] / df['Number of Employees']

# Save engineered dataset
df.to_csv("../data/Engineered_Financial_Statements.csv", index=False)

print("✅ Feature engineered dataset saved successfully!")
```
